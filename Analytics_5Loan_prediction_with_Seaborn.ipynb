```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=df)
```

```python
sns.relplot(x='Property_Area',y='LoanAmount',data=df)
```

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',hue='Married',data=df)
```

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',hue='Education',data=df)   # two numerical ,hue categorical data
```

```python
sns.catplot(x="Education",y="LoanAmount",data=df)  #one categorical and one numerical
```

```python
sns.catplot(x="Married",y="LoanAmount",data=df)
```

```python
sns.catplot(x="Loan_Status",y="ApplicantIncome",data=df)
```

```python
sns.catplot(x="Loan_Status",y="ApplicantIncome",jitter=False,data=df)
```

```python
sns.catplot(x="Education",y="LoanAmount",hue="Married",data=df)  #two categorical and one numerical
```

```python
sns.catplot(x="Education",y="LoanAmount",kind="swarm",data=df)
```

```python
sns.catplot(x="Education",y="LoanAmount",hue="Married",kind="swarm",data=df)
```

## Box plot will be helpful when there are outliers

```python
sns.catplot(x="Education",y="LoanAmount",hue="Married",kind="box",data=df)
```

```python
sns.catplot(x="Education",y="LoanAmount",hue="Married",kind="violin",data=df)
```

```python
sns.distplot(df.LoanAmount)   ##Only one numerical data
```

```python
sns.distplot(df.LoanAmount,kde=False) 
```

```python
sns.jointplot(x="LoanAmount",y="ApplicantIncome",data=df)
```

```python
sns.jointplot(x="LoanAmount",y="ApplicantIncome",kind='hex',data=df)
```

```python
sns.jointplot(x="LoanAmount",y="ApplicantIncome",kind='kde',data=df)
```
