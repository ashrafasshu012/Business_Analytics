## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Draw a line
xpoints=np.array([10,20,30])
ypoints=np.array([50,100,150])
plt.plot(xpoints,ypoints,linewidth=10)
plt.show()
```

## multiple plot

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([30,80,10,100])
plt.subplot(1,2,1)   #(rows,columns, plot number)
plt.plot(x,y)
plt.title("sales")

#plot2
x=np.array([0,1,2,3])
y=np.array([100,200,300,400])

plt.subplot(1,2,2)   #(rows,columns, plot number)
plt.plot(x,y)
plt.title("Income")
plt.show()
```

## Scatter plots(when you want to find out relationship between two numerical attributes)

```python
x=np.array([0,1,2,3])
y=np.array([5,10,15,20])
plt.scatter(x,y)
```

```python
x=np.array([0,1,2,3])
y=np.array([5,10,15,20])
plt.scatter(x,y)
x=np.array([0,1,2,3])
y=np.array([30,60,90,100])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([5,10,15,20])
plt.scatter(x,y,color='green')
x=np.array([0,1,2,3])
y=np.array([30,60,90,100])
plt.scatter(x,y,color='red')
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([5,10,15,20])
colors=(['red','yellow','black','blue'])
size=([10,200,30,40])
plt.scatter(x,y,c=colors,s=size)
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([5,10,15,20])
size=([10,200,30,40])
plt.scatter(x,y,s=size,alpha=0.6,color='r')#alpha=transparency
plt.show()
```

```python
y=np.array([10,125,60,70])
plt.pie(y)
plt.show()
```

```python
y=np.array([10,125,60,70])
mylabels=['cherries','bananas','apples','dates']
plt.pie(y,labels=mylabels)
plt.show()
```

```python
y=np.array([10,125,60,70])
mylabels=['cherries','bananas','apples','dates']
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
y=np.array([10,125,60,70])
mylabels=['cherries','bananas','apples','dates']
myexplode=[0,0.5,0.1,0.2]

plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
y=np.array([10,125,60,70])
mylabels=['cherries','bananas','apples','dates']
mycolors=['red','blue','black','green']
plt.pie(y,labels=mylabels,colors=mycolors)

plt.show()
```

```python
y=np.array([10,125,60,70])
mylabels=['cherries','bananas','apples','dates']
mycolors=['red','blue','black','green']
plt.pie(y)
plt.legend(labels=mylabels)
plt.show()
```
